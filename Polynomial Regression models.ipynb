```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split, learning_curve
%matplotlib inline
```

```python
X = np.random.rand(200) * 18
Y = 0.36 * X**4 - 0.6 * X**3 + 0.46 * X**2 +0.81 * X + 1
Y = Y + np.random.normal(scale = 4000, size = Y.shape)
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)
plt.plot(X,Y, 'bo')
MSE = []
MSET = []
```

```python
split = int(X.shape[0] * 0.7)
X_train = X[:split]
Y_train = Y[:split]
X_test = X[split:]
Y_test = Y[split:]
```

```python
model = LinearRegression(normalize = True)
poly1 = PolynomialFeatures(degree = 1)
X1 = poly1.fit_transform(X_train)
X1_t = poly1.fit_transform(X_test)
model.fit(X1,Y_train)
```

```python
X1_plot = np.arange(-20,20,0.01).reshape(-1,1)
X1_plot_poly = poly1.fit_transform(X1_plot)
plt.scatter(X_train, Y_train, color = "blue")
plt.scatter(X_test, Y_test, color = "cyan")
plt.plot(X1_plot[:,0],model.predict(X1_plot_poly),'r--')
```

```python
h1 = model.predict(X1)
MSE.append(np.mean((Y_train - h1) ** 2)/2) 
ht1 = model.predict(X1_t)
MSET.append(np.mean((Y_test - ht1) ** 2)/2) 

print('training score:', model.score(X1, Y_train))
print('test score:', model.score(X1_t, Y_test))
```
